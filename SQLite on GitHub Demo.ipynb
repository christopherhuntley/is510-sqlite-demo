# SQLite & Jupyter on GitHub Demo
__This demonstration shows how SELECT queries can be run against a read-only database in GitHub.__

```python
import sqlite3
import pandas as pd

%load_ext sql
%sql sqlite:///deals.db
```

```python
%%sql 
SELECT * from Deals;
```

Now lets try to write data to the `deals.db` file. 

```python
%%sql
UPDATE Deals SET DealName = 'pwned' WHERE DealID=1; 
SELECT * from Deals WHERE DealID=1;
```
